```python
# import potrzebnych bibliotek
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

## Zadanie 3.

```python
y = lambda x: x ** 2 + 5  # obliczanie funnkcji dla jednej wartości x

# tworzenie przedziałów argumentów funkcji
x_range_1 = np.linspace(-1, 1, 100)
x_range_2 = np.linspace(-6, 6, 100)
x_range_3 = np.linspace(0, 5, 100)

# obliczanie wartości funkcji w danych przedziałach
y_range_1 = [y(x) for x in x_range_1]
y_range_2 = [y(x) for x in x_range_2]
y_range_3 = [y(x) for x in x_range_3]
```

```python
# wyświetlenie pierwszego przedziału
plt.figure()
plt.plot(x_range_1, y_range_1)
plt.title('Przedział pierwszy funkcji')
plt.ylabel('Wartości funkcji')
plt.xlabel('Argumenty')
plt.legend(['y=x^2+5'])
plt.show()
```

```python
# wyświetlenie drugiego przedziału
plt.figure()
plt.plot(x_range_2, y_range_2)
plt.title('Przedział drugi funkcji')
plt.ylabel('Wartości funkcji')
plt.xlabel('Argumenty')
plt.legend(['y=x^2+5'])
plt.show()
```

```python
# wyświetlenie trzeciego przedziału
plt.figure()
plt.plot(x_range_3, y_range_3)
plt.title('Przedział trzeci funkcji')
plt.ylabel('Wartości funkcji')
plt.xlabel('Argumenty')
plt.legend(['y=x^2+5'])
plt.show()
```
